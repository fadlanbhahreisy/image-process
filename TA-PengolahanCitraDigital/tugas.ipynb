```python
import cv2         
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image = cv2.imread('foto.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
print("Pixel Ori")
print(image)
```

```python
print("Original Image")
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
print("pixel Grayscale")
print(gray)
```

```python
gray_image = cv2.cvtColor(gray, cv2.COLOR_BGR2RGB)
print("Grayscale Image")
plt.imshow(gray_image)
```

```python
ret, thresh1 = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY + 
                                            cv2.THRESH_OTSU)   
```

```python
thresh = cv2.cvtColor(thresh1, cv2.COLOR_BGR2RGB)
```

```python
print("Pixel Threshold Hitam Putih")
print(thresh1)
```

```python
print("Gambar Hitam Putih hasil Threshold Otsu")
plt.imshow(thresh)
```
